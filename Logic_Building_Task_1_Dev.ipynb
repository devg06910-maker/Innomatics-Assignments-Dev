```python
# Task: Credentials given, check whether login successful or not

# Correct credentials
username = "admin"
password = "1234"

# User input
u = input("Enter username: ")
p = input("Enter password: ")

# Check login
if u == username and p == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Task: Analyze the given student results

# List of marks
marks = [45, 78, 90, 33, 60]

pass_count = 0   # Counter for pass
fail_count = 0   # Counter for fail

# Check each mark
for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print result
print("Pass:", pass_count)
print("Fail:", fail_count)
```

```python
# Task: Clean and standardize the user names

# Names with spaces and mixed case
names = [" Alice ", "bob", " CHARLIE "]

# Clean each name
for i in range(len(names)):
    names[i] = names[i].strip().lower()   # remove spaces + lowercase

print(names)
```

```python
# Task: Analyze message sizes

# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Loop through each message
for msg in messages:
    print("Message:", msg)

    # Calculate length using len()
    length = len(msg)
    print("Length:", length)

    # Check if message is longer than 10 characters
    if length > 10:
        print("This message is longer than 10 characters")

    print()  # Blank line for better output formatting
```

```python
# Task: Detection of error messages from system logs

# System logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

count = 0   # Error counter

# Count ERROR entries
for l in logs:
    if l == "ERROR":
        count += 1

print("Errors:", count)
```
