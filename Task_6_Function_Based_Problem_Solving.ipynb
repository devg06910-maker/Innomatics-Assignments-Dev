```python
# Problem 1: Smart Parking Lot Management System

def manage_parking(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            current -= 1

        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)

    if current > capacity:
        print("Parking Status: Full - Capacity Exceeded")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)

capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(capacity, logs)
```

```python
# Problem 2: Online Food Delivery Time Estimator

def estimate_delivery(distance, traffic, weather):
    eta = distance * 5   # 5 minutes per km base time

    if traffic == "High":
        eta += 15
    elif traffic == "Medium":
        eta += 8

    if weather == "Rainy":
        eta += 10
    elif weather == "Storm":
        eta += 20

    print("Estimated Delivery Time:", eta, "minutes")

estimate_delivery(8, "High", "Rainy")

```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer

def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", occupancy, "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

total_seats = 200
booked_seats = [1] * 150   # 150 bookings

analyze_occupancy(total_seats, booked_seats)
```

```python
# Problem 4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    print("Average CPU Load:", average, "%")

    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Server Status:", status)

cpu = [45, 60, 70, 85, 90]
classify_server_load(cpu)
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor

def monitor_resources(resources):
    overused = []

    for resource in resources:
        if resources[resource] > 8:
            overused.append(resource)

    print("Overused Resources:", overused)

    if len(overused) > 0:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage)
```

```python
# Problem 6: Online Event Registration Capacity Controller

def manage_registration(capacity, registrations):

    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

manage_registration(100, 105)
```
