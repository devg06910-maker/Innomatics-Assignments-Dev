```python
# Problem 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest = max(employees.values())

# Store employees who have highest score
top = []

for name in employees:
    if employees[name] == highest:
        top.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top), "(Score:", highest, ")")
```

```python
# Problem 2: Search Query Keyword Analysis

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Split into words
words = query.split()

freq = {}

# Count word frequency
for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

# Print words repeated more than once
result = {}

for word in freq:
    if freq[word] > 1:
        result[word] = freq[word]

print(result)
```

```python
# Problem 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid = []

# Check each reading
for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:   # Even number check
        valid.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total = len(emails)

# Calculate percentage
for domain in domain_count:
    percent = (domain_count[domain] / total) * 100
    print(domain + ":", str(int(percent)) + "%")
```

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
average = sum(sales) / len(sales)

# Check for spikes (more than 30% above average)
for i in range(len(sales)):
    if sales[i] > average * 1.3:
        print("Day", i + 1, ":", sales[i])
```

```python
# Problem 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count = {}

# Count occurrences
for user in user_ids:
    if user in count:
        count[user] += 1
    else:
        count[user] = 1

# Print duplicates only
for user in count:
    if count[user] > 1:
        print(user, "→", count[user], "times")
```
